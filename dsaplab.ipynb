```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
```

```python
#Continuous Sin Wave
a = 5
f = 3
t =np.arange(0, 1, 0.001)
x = a * np.sin(2*np.pi*f*t)
plt.plot(t, x)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('CSin/Abiralbhattarai/10')
plt.grid(True)
plt.show()
```

```python
#Continuous Cosine Wave
a = 5
f = 3
t =np.arange(0, 1, 0.001)
x = a * np.cos(2*np.pi*f*t)
plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('CCos/Abiralbhattarai/10')
plt.grid(True)
plt.show()
```

```python
#Discrete Sin Wave
a = 5
f = 3
t =np.arange(0, 1, 0.01)
x = a * np.sin(2*np.pi*f*t)
plt.stem(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('DSin/Abiralbhattarai/10')
plt.grid(True)
plt.show()
```

```python
#Discrete Cosine Wave
a = 5
f = 3
t =np.arange(0, 1, 0.01)
x = a * np.cos(2*np.pi*f*t)
plt.stem(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('DCos/Abiralbhattarai/10')
plt.grid(True)
plt.show()
```

```python
#Subplot with continuous and discrete sin & cosine waves
a=5
f=4
t = np.arange(0,1,0.01)
x = a*np.sin(2*np.pi*f*t)
y = a*np.cos(2*np.pi*f*t)
plt.subplot(2,2,1)
plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('CSin/Abiralbhattarai/10')
plt.grid(True)
plt.subplot(2,2,2)
plt.plot(t, y)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('CCos/Abiralbhattarai/10')
plt.grid(True)
plt.subplot(2,2,3)
plt.stem(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('DSin/Abiralbhattarai/10')
plt.grid(True)
plt.subplot(2,2,4)
plt.stem(t, y)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('DCos/Abiralbhattarai/10')
plt.grid(True)
plt.show()
```

```python
#Square Wave
a=5
f=4
t = np.arange(0,1,0.01)
x = a * signal.square(2 * np.pi * f * t)
plt.plot(t, x)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Square Wave/AbiralBhattarai/10')
plt.grid(True)
plt.show()
```
